##### Initialise Project Modules

```python
#backend modules
from configuration import ConfigurationFactory
from service.registry import RegistryManager
from service.log import LogService
from model.dbmodel import DatabaseService, ShortEndDatabaseManager
from analysis.analytics import DataAnalyticsService
from matplotlib.backends.backend_gtk3agg import FigureCanvasGTK3Agg as FigureCanvas

#front end modules
%matplotlib inline
from model.payload import Payload, PayloadProcessor, payloads
from graph.graph import Lineplot, Scatterplot, Heatplot
```

```python
#configuration
config = ConfigurationFactory.create_config()

#initialise all of the services
log_serv = LogService(config["system"]["log"])
db_serv = DatabaseService(config["system"]["db"])
analytics = DataAnalyticsService()

#start all services
log_serv.start()
db_serv.start()
analytics.start()
```

##### Process analysis payloads and create datamodel

```python
#process payload with server and create datamodel (temp)
datamodel = {}
for plid, pl in payloads.items():
    PayloadProcessor(pl, db_serv.db)
    datamodel[plid] = analytics.gen_plot_datamodel(payloads[plid])
    
stl_hp = analytics.gen_heatplot_matrix(payloads["p1"])
ebor_hp = analytics.gen_heatplot_matrix(payloads["p4"])
```

### Euribor (1Y) Vol Summary

Commentary:

- Over the past year this is the first time we are seeing vol and futures at these levels across the entire curve (I, K, K2 and K3)
- 97% of the realized vol in K Q1 currently below the implied vol level (most expensive on the heatmap)
- The implied vol is slightly more expensive in 1YR midcurve ebor than it is in the greens (1y relative comparison from the heatmap)
